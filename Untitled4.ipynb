Résumé des options de personnalisation de l’interface de la courbe :

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(10, 20, 2)
y = 5*x + 6
# Insérer une légende :
plt.plot(x, y, label = "Légende de la courbe")
plt.legend()
# limiter les bornes d'affichage selon les axes :
plt.xlim(13, 18)
plt.ylim(70, 100)
# Insérer des sous-titres aux axes :
plt.xlabel("axe des abscisses")
plt.ylabel("axe des ordonnées")
# Insérer un titre :
plt.title("Titre")
# Afficher la grille :
plt.grid()
# Afficher la courbe :
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(10, 20, 2)
y1 = 2*x + 10
y2 = 3*x
plt.plot(x, y1, 'y')
plt.plot(x, y2, 'r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x, y1, "x")
plt.plot(x, y2, '.')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x, y1, linewidth = 10)
plt.plot(x, y2, linewidth = 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2, 7, 100)
y1 = 2*np.cos(x)
y2 = np.sin(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.text(0, 2, "le maximum de la fonction 2cos(x)")
plt.text(2, 1, "le maximum de la fonction sin(x)")
plt.show()
```
